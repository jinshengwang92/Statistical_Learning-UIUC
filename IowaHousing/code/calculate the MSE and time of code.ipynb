```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# LS all
MSE_ls = [0.01537976,0.0231954,0.01910991,0.02428723,0.02011824]
time_ls = [1.485793,1.216253,1.217694,1.312877,1.330241]
mean_MSE_LS = np.mean(MSE_ls)
std_MSE_LS = np.std(MSE_ls)
mean_time_LS = np.mean(time_ls)
std_time_LS = np.std(time_ls)
print('For Simple liear regression, MSE mean: {0:.4f}, std: {1:.4f}; Time mean: {2:.4f}, std: {3:.4f}'\
      .format(mean_MSE_LS,std_MSE_LS,mean_time_LS,std_time_LS))
```

```python
# lasso
MSE_lasso = [0.01339167,0.01994688,0.01638137,0.02089965,0.01688194]
time_la = [1.247198,1.312788,1.271789,1.246561,1.128076]
mean_MSE_lasso = np.mean(MSE_lasso)
std_MSE_lasso = np.std(MSE_lasso)
mean_time_lasso = np.mean(time_la)
std_time_lasso = np.std(time_la)
print('For Lasso, MSE mean: {0:.4f}, std: {1:.4f}; Time mean: {2:.4f}, std: {3:.4f}'\
      .format(mean_MSE_lasso,std_MSE_lasso,mean_time_lasso,std_time_lasso))
```

```python
# rf
MSE_rf = [0.01504062,0.02007721,0.01808417,0.02102912,0.02213319]
time_f =[26.97315,25.0889,25.59084,25.24171,25.47824]
mean_MSE_rf = np.mean(MSE_rf)
std_MSE_rf = np.std(MSE_rf)
mean_time_rf = np.mean(time_f)
std_time_rf = np.std(time_f)
print('For Random Forest, MSE mean: {0:.4f}, std: {1:.4f}; Time mean: {2:.4f}, std: {3:.4f}'\
      .format(mean_MSE_rf,std_MSE_rf,mean_time_rf,std_time_rf))
```

```python
# data processing time 0.219s
time_pre_prosess = [0.2281199,0.197818,0.1877801,0.1930151,0.1884999]
mean_time_pre = np.mean(time_pre_prosess)
std_time_pre = np.std(time_pre_prosess)
print('Data preprocessing time,mean: {0:.4f}, std: {1:.4f}'\
      .format(mean_time_pre,std_time_pre))
```

```python
#fig,ax =plt.subplots()
x = [1,2,3,4,5]
plt.plot(x,MSE_ls,'r*',label='Simple Linear({0:.4f}+-{1:.4f})'.format(mean_MSE_LS,std_MSE_LS),linestyle='-')
plt.plot(x,MSE_lasso,'kD',label='Lasso Regression({0:.4f}+-{1:.4f})'.format(mean_MSE_lasso,std_MSE_lasso),linestyle='-')
plt.plot(x,MSE_rf,'b^',label='Random Forest({0:.4f}+-{1:.4f})'.format(mean_MSE_rf,std_MSE_rf),linestyle='-')
ax = plt.gca()
ax.set_xlabel('# of trial')
ax.set_ylabel('Mean square error')
ax.set_title('Mean Square Error of three methods')
plt.legend(loc='best')
plt.savefig('MSE_3_methods.pdf')
plt.show()
```
